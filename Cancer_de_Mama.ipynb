### Transformações utilizando a clausula CASE

```python
consulta_2 = """
SELECT DISTINCT
	CLASSE
FROM
	DATASET



"""
#Com essa consulta temos os valores únicos da coluna 'Classe'. Como o resultado foi somemente de 2 valores, iremos aplicar a binarização.
pd.read_sql(consulta_2, cnn)
```

```python
#BINARIZACAO DA VARIAVEL 'Classe'
consulta_3 = """
SELECT
	CASE
		WHEN CLASSE = 'no-recurrence-events' THEN 0
        WHEN CLASSE = 'recurrence-events' THEN 1
	END AS classe
FROM
	DATASET

"""
pd.read_sql(consulta_3, cnn)
```

```python
consulta_4 = """
SELECT DISTINCT
	IRRADIANDO
FROM
	DATASET



"""
#Com essa consulta temos os valores únicos da coluna 'Irradiando'. Como o resultado foi somemente de 2 valores, iremos aplicar a binarização.
pd.read_sql(consulta_4, cnn)
```

```python
#BINARIZACAO DA VARIAVEL 'Irradiando'
consulta_5 = """
SELECT
	CASE
		WHEN IRRADIANDO = 'no' THEN 0
		WHEN IRRADIANDO = 'yes' THEN 1
	END AS irradiando
FROM
	DATASET

"""
pd.read_sql(consulta_5, cnn)
```

```python
consulta_6 = """
SELECT DISTINCT
	NODE_CAPS
FROM
	DATASET

"""
#Com essa consulta temos os valores únicos da coluna 'Irradiando'. Como o resultado foi mais que 2 valores, iremos aplicar a codificação.
pd.read_sql(consulta_6, cnn)
```

```python
#CODIFICACAO DA VARIAVEL 'Node_caps'
consulta_7 = """
SELECT
	CASE
		WHEN NODE_CAPS = 'no' THEN 0
        WHEN NODE_CAPS = 'yes' THEN 1
        ELSE 2
	END AS node_caps
FROM
	DATASET

"""
#Vale ressaltar que os valores 0 se referem a 'no' , 1 a 'yes' e tudo que difere deles, que podem ser valores NaN, estarão como 2.
pd.read_sql(consulta_7, cnn)
```

```python
consulta_8 = """
SELECT DISTINCT
	SEIO
FROM
	DATASET

"""
#Com essa consulta temos os valores únicos da coluna 'seio'. Como o resultado foi somemente de 2 valores, iremos aplicar a binarização.
pd.read_sql(consulta_8, cnn)
```

```python
#BINARIZACAO DA VARIAVEL 'seio'
consulta_8 = """
SELECT
	CASE
		WHEN SEIO = 'left' THEN 0
        WHEN SEIO = 'right' THEN 1
	END AS seio
FROM
	DATASET

"""

pd.read_sql(consulta_8, cnn)
```

```python
consulta_9 = """
SELECT DISTINCT
	TAMANHO_TUMOR
FROM
	DATASET
ORDER BY
	TAMANHO_TUMOR

"""
#Com essa consulta temos os valores únicos da coluna 'tamanho_tumor'. Como o resultado foi diversos valores unicos, iremos aplicar a categorização. Assim, fazendo
#o agrupamento dos valores mediante uma regra de négocio ja estabelecida.

#Iremos adotar a seguinte regra:

#Valores entre 0 á 9 = 'PP'
#Valores entre 10 á 14 = 'P'
#Valores entre 20 á 29 = 'M'
#Valores entre 30 á 39 = 'G'
#Valores entre 40 á 49 = 'GG'
#Valores acima de 49 = 'XGG'
pd.read_sql(consulta_9, cnn)

```

```python
#CATEGORIZACAO DA VARIAVEL 'tamanho_tumor'
consulta_10 = """
SELECT
	CASE
		WHEN  TAMANHO_TUMOR = '0-4' OR TAMANHO_TUMOR = '5-9' THEN 'PP'
		WHEN  TAMANHO_TUMOR = '10-14' OR TAMANHO_TUMOR = '15-19' THEN 'P'
        WHEN  TAMANHO_TUMOR = '20-24' OR TAMANHO_TUMOR = '25-29' THEN 'M'
        WHEN  TAMANHO_TUMOR = '30-34' OR TAMANHO_TUMOR = '35-39' THEN 'G'
        WHEN  TAMANHO_TUMOR = '40-44' OR TAMANHO_TUMOR = '45-49' THEN 'GG'
        WHEN  TAMANHO_TUMOR = '50-54' THEN 'XGG'
	END AS tamanho_tumor
FROM
	DATASET

"""

pd.read_sql(consulta_10, cnn)
```

```python
consulta_11 = """
SELECT DISTINCT
	QUADRANTE
FROM
	DATASET


"""
#Com essa consulta temos os valores únicos da coluna 'quadrante'. Como o resultado foi diversos valores unicos, iremos aplicar a codificação. 
#Iremos adotar a seguinte regra:

#	'left_low' = 1
#'right_up' = 2
#'left_up' = 3
#'right_low' = 4
#'central' = 5
# qualquer outro valor = 0
pd.read_sql(consulta_11, cnn)
```

```python
#CODIFICACAO DA VARIAVEL 'quadrante' (1,2,3,4,5) {0 REPRESENTA VALOR NULO}
consulta_12 = """
SELECT
	CASE
		WHEN QUADRANTE = 'left_low' THEN 1
        WHEN QUADRANTE = 'right_up' THEN 2
		WHEN QUADRANTE = 'left_up' THEN 3
        WHEN QUADRANTE = 'right_low' THEN 4
        WHEN QUADRANTE = 'central' THEN 5
        ELSE 0
	END AS QUADRANTE
FROM
	DATASET
"""
pd.read_sql(consulta_12, cnn)
```

###Criando Um Novo Dataset

```python
#Criando a tabela que recebera o novo dataset
consulta_13 = """
CREATE TABLE dataset_novo (classe INT(2),
                        idade VARCHAR (10),
                        menopausa VARCHAR (10),
                        tamanho_tumor VARCHAR (10),
                        inv_nodes VARCHAR (10),
                        node_caps INT(2),
                        deg_malig INT(2),
                        seio VARCHAR (10),
			                  quadrante INT(2),
                        irradiando INT(2))
"""
pd.read_sql(consulta_13, cnn)
```

```python
#Visualizando  a tabela criada
consulta_14 = """
SELECT
  *
FROM
  dataset_novo
"""
pd.read_sql(consulta_14, cnn)
```

```python
#Feito todos esses tratamentos, iremos criar um novo dataset aplicando todas essas transformações
consulta_15 = """

INSERT INTO dataset_novo (classe ,
                        idade ,
                        menopausa ,
                        tamanho_tumor ,
                        inv_nodes ,
                        node_caps ,
                        deg_malig ,
                        seio ,
			                  quadrante ,
                        irradiando )
SELECT
	CASE
		WHEN classe = 'no-recurrence-events' THEN 0
    WHEN classe = 'recurrence-events' THEN 1
	END AS classe,
idade,
menopausa,
	CASE
		WHEN  tamanho_tumor = '0-4' OR tamanho_tumor = '5-9' THEN 'PP'
		WHEN  tamanho_tumor = '10-14' OR tamanho_tumor = '15-19' THEN 'P'
    WHEN  tamanho_tumor = '20-24' OR tamanho_tumor = '25-29' THEN 'M'
    WHEN  tamanho_tumor = '30-34' OR tamanho_tumor = '35-39' THEN 'G'
    WHEN  tamanho_tumor = '40-44' OR tamanho_tumor = '45-49' THEN 'GG'
    WHEN  tamanho_tumor = '50-54' THEN 'XGG'
	END AS tamanho_tumor,
inv_nodes,
  CASE
		WHEN node_caps = 'no' THEN 0
    WHEN node_caps = 'yes' THEN 1
    ELSE 2
	END AS node_caps, 
deg_malig,
  CASE
		WHEN seio = 'left' THEN '0'
    WHEN seio = 'right' THEN '1'
	END AS seio,
  CASE
	  WHEN quadrante = 'left_low' THEN 1
    WHEN quadrante = 'right_up' THEN 2
		WHEN quadrante = 'left_up' THEN 3
    WHEN quadrante = 'right_low' THEN 4
    WHEN quadrante = 'central' THEN 5
    ELSE 0
	END AS quadrante,
    
	CASE
		WHEN irradiando = 'no' THEN 0
		WHEN irradiando = 'yes' THEN 1
	END AS irradiando
  
FROM
	dataset

"""
pd.read_sql(consulta_15, cnn)
```
